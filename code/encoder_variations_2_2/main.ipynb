# Encoder Variation Runner

Run this notebook to execute the encoder experiment notebooks in order:
1. DeBERTa
2. RemBERT
3. XLM-RoBERTa


```python
from pathlib import Path
import subprocess
import sys

NOTEBOOKS = [
    "Q2_2_DeBERTa_Experiments_with_output.ipynb",
    "Q2_2_RemBERT_Experiments_with_output.ipynb",
    "XLMRoBERTa_Sentiment_Experiments.ipynb",
]

def resolve_notebook_dir() -> Path:
    candidates = [Path.cwd(), Path.cwd() / "encoder_variations_2_2"]
    for base in candidates:
        if all((base / nb).exists() for nb in NOTEBOOKS):
            return base
    raise FileNotFoundError("Could not locate encoder_variations_2_2 notebook directory.")

base_dir = resolve_notebook_dir()
print(f"Using notebook directory: {base_dir}")

for notebook_name in NOTEBOOKS:
    notebook_path = base_dir / notebook_name
    print(f"\nRunning: {notebook_name}")
    cmd = [
        sys.executable,
        "-m",
        "jupyter",
        "nbconvert",
        "--to",
        "notebook",
        "--inplace",
        "--execute",
        str(notebook_path),
    ]
    subprocess.run(cmd, check=True)

print("\nAll notebooks completed in order.")
```
